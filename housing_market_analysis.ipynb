```python
#Remove outliers using interquartile range, keep 50% data between 25% to 75% data
q1 = df['Price'].quantile(0.25)
q3 = df['Price'].quantile(0.75)

print(q1, q3)

iqr = q3 - q1

print(iqr)

lower_boundary = q1 - 1.5*iqr
upper_boundary = q3 + 1.5*iqr
print(lower_boundary, upper_boundary)

removed_outliers = df[(df['Price'] > lower_boundary) & df['Price'] <= upper_boundary]
removed_outliers
```
