```python
#Lets filter the houses with 3 bedrooms and atleast 2000sqft area
filter_beds_area =  df[(df['Bedrooms']==3) & (df['Area'] <= 2000)]
print(filter_beds_area.head())
```

```python
#Sorting the data for easier analysis and trend
sorted_df = df.sort_values('Price')
print(sorted_df[['Price','Bedrooms','Area']])
```

```python
#Sort by largest area and see price
sorted_area_df = df.sort_values(by ='Area', ascending= False)
print(sorted_area_df[['Area','Price','Distance_to_CBD']])
```

```python
#Average and Grouping
#Average price by number of bedrooms
avg_per_bedrooms = df.groupby('Bedrooms')['Price'].mean()
print(avg_per_bedrooms)
```

```python
#Lets count number of houses with certain numbers of bedrooms
house_bedrooms = df['Bedrooms'].value_counts()
print(house_bedrooms)

# Q. How can we also add average price per number of bedrooms
# house_bedrooms = df['Bedrooms'].value_counts().df['Price'].mean
```
